# Steps for Data Preparation
## 1. Diminish dimensionality by taking out number of attributes that are irrelevant for each NBA player
## 2. Find all stats for each NBA player over the last 3 seasons and combine into one row for each player.
## 3. After that use data analysis to find and compare each NBA player.

```python
#dropping irrelevant columns that I do not need
NBADROP = NBA.drop(['mean_views', 'Fvot', 'FRank', 'Mvot', 'MRank' , 'Score' , 'Play' , 'Pvot', 'PRank', 'Conference'], axis = 1)
```

```python
#Make an Array to find all the averages of each of the NBA player over 3 seasons!
Player_ID = NBADROP.Player_ID.unique()
Player_ID.__len__()
#Found out how many player were in the league for 3 seasons
```

```python
#Make a new data frame to add each player in data set and their averages over three seasons
NBAAVERAGES = pd.DataFrame()
for player in Player_ID:
    #Provided averages over 3 seasons, and calculated total games played
    ptsavg = NBADROP.loc[NBADROP['Player_ID'] == player, 'PTS'].sum()/3
    astsavg = NBADROP.loc[NBADROP['Player_ID'] == player, 'AST'].sum()/3
    rebsavg = NBADROP.loc[NBADROP['Player_ID'] == player, 'TRB'].sum()/3
    stlsavg = NBADROP.loc[NBADROP['Player_ID'] == player, 'STL'].sum()/3
    blksavg = NBADROP.loc[NBADROP['Player_ID'] == player, 'BLK'].sum()/3
    orebavg = NBADROP.loc[NBADROP['Player_ID'] == player, 'ORB'].sum()
    drebsavg = NBADROP.loc[NBADROP['Player_ID'] == player, 'DRB'].sum()/3
    gamesplayed = NBADROP.loc[NBADROP['Player_ID'] == player, 'G'].sum()
    gamesstarted = NBADROP.loc[NBADROP['Player_ID'] == player, 'GS'].sum()
    fieldgoal= NBADROP.loc[NBADROP['Player_ID'] == player, 'FG'].sum()/3
    attemptsfieldgoal  = NBADROP.loc[NBADROP['Player_ID'] == player, 'FGA'].sum()/3
    Minutes = NBADROP.loc[NBADROP['Player_ID'] == player, 'MP'].sum()/3
    personalfouls = NBADROP.loc[NBADROP['Player_ID'] == player, 'PF'].sum()/3
    FGP = NBADROP.loc[NBADROP['Player_ID'] == player, 'FG.'].sum()/3
    X2P = NBADROP.loc[NBADROP['Player_ID'] == player, 'X2P'].sum()/3
    X3P = NBADROP.loc[NBADROP['Player_ID'] == player, 'X3P'].sum()/3
    X3PA = NBADROP.loc[NBADROP['Player_ID'] == player, 'X3PA'].sum()/3
    X2PA = NBADROP.loc[NBADROP['Player_ID'] == player, 'X2PA'].sum()/3
    X3PP = NBADROP.loc[NBADROP['Player_ID'] == player, 'X3P.'].sum()/3
    X2PP = NBADROP.loc[NBADROP['Player_ID'] == player, 'X2P.'].sum()/3
    FT = NBADROP.loc[NBADROP['Player_ID'] == player, 'FT'].sum()/3
    FTA = NBADROP.loc[NBADROP['Player_ID'] == player, 'FTA'].sum()/3
    FTP = NBADROP.loc[NBADROP['Player_ID'] == player, 'FT.'].sum()/3
    EFG = NBADROP.loc[NBADROP['Player_ID'] == player, 'eFG.'].sum()/3
    Role = NBADROP.loc[NBADROP['Player_ID'] == player, 'Role'].values[0]
    #Added all 660 players to new Data Frame
    NBAAVERAGES = NBAAVERAGES.append({'Player ID': player, 'Name': NBADROP.loc[NBADROP['Player_ID'] == player, 'Player.x'].values[0],'Points Per Game': round(ptsavg, 1), 'Assists per game': round(astsavg, 1),
                                      'Total Rebounds Per Game': round(rebsavg, 1), 'Steals Per Game': round(stlsavg, 1),
                                'Offensive Rebounds Per Game': round(orebavg, 1), 'Defensive Rebounds Per Game': round(drebsavg, 1) 
                                , 'Blocks Per Game': round(blksavg,1), 'Field Goals Made': round(fieldgoal, 1), 
                                 'Field Goal Attempts':round(attemptsfieldgoal,1),'Field Goal Percentage': round(FGP, 3)*100, 
                                      'Effective Field Goal Percentage %': round(EFG, 3) * 100,
                                 '2-Pointers Made':round(X2P, 1), '2-Pointers Attempted' : round(X2PA, 1), '2-POINT PERCENTAGE %': round(X2PP, 3)*100,
                                 '3-Pointers Made': round(X3P, 1), '3-Pointers Attempted':round(X3PA, 1), '3-POINT PERCENTAGE %': round(X3PP, 3) * 100,
                                 'FREE-THROWS MADE': round(FT, 1), 'FREE THROWS ATTEMPTED':round(FTA, 1), 'FREE THROW PERCENTAGE %': round(FTP, 3) * 100,
                                 'Personal Fouls Per Game': round(personalfouls, 1),
                                 'Minutes Per Game': round(Minutes, 1), 'GAMES STARTED': gamesstarted, 'GAMES PLAYED': gamesplayed,
                                 'Position': NBADROP.loc[NBADROP['Player_ID'] == player, 'Pos1'].values[0], 
                                 'Salary': round(NBADROP.loc[NBADROP['Player_ID'] == player, 'Salary'].sum()/3, 6),
                                 'Role': Role}, ignore_index=True)
```

Let us see if the two new variables are highly correlated using linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn import metrics
X_nba = NBAAVERAGES['DEFENSIVE STATS'].values.reshape(-1,1)
y_nba = NBAAVERAGES['OFFENSIVE STATS'].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X_nba, y_nba, test_size=0.2, random_state=0)
model = LinearRegression()
data = model.fit(X_train,y_train)
print(data.coef_)
print(data.intercept_)
print(data.score(X_nba, y_nba))
y_pred = model.predict(X_test)
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
plt.scatter(X_test, y_test,  color='gray')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()
df
```

From this linear regression model we can see that the model is predicting properly. This shows that a player needs to be able to play defense and offense to impact winning. So the model is predicting that the increase in defensive production and offensive production leads to becoming a better player over all which leads to being a top player on the team.
